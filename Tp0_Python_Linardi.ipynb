```python
import matplotlib.pyplot as plt
import numpy as np

# Señal cuadrada (escalón)
senial_cuad = [] #inicializo un array vacío
for a in range(5):
    senial_cuad.append(0)
for a in range(5):
    senial_cuad.append(1)
print(senial_cuad)
plt.title("Escalón")
plt.plot(senial_cuad)
plt.show()

##### MAXIMOS Y MINIMOS ESCALON #########

# Inicializa valores y listas de índices

minimo_valor = senial_cuad[0]
maximo_valor = senial_cuad[0]
indices_minimos = []
indices_maximos = []

# Recorre la lista para encontrar mínimos y máximos

for i in range(len(senial_cuad)):
    if senial_cuad[i] < minimo_valor:
        minimo_valor = senial_cuad[i]
        indices_minimos = [i]  # Nuevo indice minimo
    elif senial_cuad[i] == minimo_valor:
        indices_minimos.append(i)  # Agrego índice si es igual al mínimo

    if senial_cuad[i] > maximo_valor:
        maximo_valor = senial_cuad[i]
        indices_maximos = [i]  # Nuevo indice maximo
    elif senial_cuad[i] == maximo_valor:
        indices_maximos.append(i)  # Agrega índice si es igual al máximo

# Muestra resultados

print("Valor mínimo:", minimo_valor)
print("Índices del valor mínimo:", indices_minimos)

print("Valor máximo:", maximo_valor)
print("Índices del valor máximo:", indices_maximos)

##############################


# Señal delta de dirac
delta_dirac = []
for a in range(100):
    delta_dirac.append(0)
dd= int(len(delta_dirac)/2) #obtener la posición del medio del array
delta_dirac[dd] = 1
print("\nel largo del vector delta_dirac es: {}".format(len(delta_dirac)))
plt.title("Dirac")
plt.plot(delta_dirac)
plt.show()

##### MAXIMOS Y MINIMOS DELTA DE DIRAC #########

# Inicializa valores y listas de índices

minimo_valor2 = delta_dirac[0]
maximo_valor2 = delta_dirac[0]
indices_minimos2 = []
indices_maximos2 = []

# Recorre la lista para encontrar mínimos y máximos

for i in range(len(delta_dirac)):
    if delta_dirac[i] < minimo_valor2:
        minimo_valor2 = delta_dirac[i]
        indices_minimos2 = [i]  # Nuevo índice mínimo
    elif delta_dirac[i] == minimo_valor2:
        indices_minimos2.append(i)  # Agrego índice si es igual al mínimo

    if delta_dirac[i] > maximo_valor2:
        maximo_valor2 = delta_dirac[i]
        indices_maximos2 = [i]  # Nuevo indice maximo
    elif delta_dirac[i] == maximo_valor2:
        indices_maximos2.append(i)  # Agrega índice si es igual al máximo

# Muestra resultados

print("Valor mínimo:", minimo_valor2)
print("Índices del valor mínimo:", indices_minimos2)

print("Valor máximo:", maximo_valor2)
print("Índices del valor máximo:", indices_maximos2)

##############################

# Señal triangular
senial_tria = []
for a in range(5):
    senial_tria.append(a)
for a in range(5,0,-1):
    senial_tria.append(a)
plt.title("Triangular")
plt.plot(senial_tria)
plt.show()

##### MAXIMOS Y MINIMOS SEÑAL TRIANGULAR #########

# Inicializa valores y listas de índices

minimo_valor3 = senial_tria[0]
maximo_valor3 = senial_tria[0]
indices_minimos3 = []
indices_maximos3 = []

# Recorre la lista para encontrar mínimos y máximos

for i in range(len(senial_tria)):
    if senial_tria[i] < minimo_valor3:
        minimo_valor3 = senial_tria[i]
        indices_minimos3 = [i]  # Nuevo índice mínimo
    elif senial_tria[i] == minimo_valor3:
        indices_minimos3.append(i)  # Agrego índice si es igual al mínimo

    if senial_tria[i] > maximo_valor3:
        maximo_valor3 = senial_tria[i]
        indices_maximos3 = [i]  # Nuevo indice maximo
    elif senial_tria[i] == maximo_valor3:
        indices_maximos3.append(i)  # Agrega índice si es igual al máximo

# Muestra resultados

print("Valor mínimo:", minimo_valor3)
print("Índices del valor mínimo:", indices_minimos3)

print("Valor máximo:", maximo_valor3)
print("Índices del valor máximo:", indices_maximos3)

##############################



#Se puede definir funciones y parametrizar las señales que voy a generar
#Ejemplo con delta de dirac
def generar_delta_dirac(cantidad,pos):
    delta_dirac = []
    for a in range(cantidad):
        delta_dirac.append(0)
    delta_dirac[pos] = 1
    return delta_dirac

print("\n")
print(generar_delta_dirac(10,5))
```

```python
# Ejemplo:  Señal cuadrada con scipy
import scipy.signal as sig
import matplotlib.pyplot as plt

#Parámetros de la señal
frec_muestreo = 5000 #en Hz
frec_senial = 2 # en Hz
# Base de tiempo de 2 formas (es indistinto cuál usar)
# t = np.linspace(0, 1, frec_muestreo, endpoint=False)
t = np.arange(0,1,1/frec_muestreo)

print("t: {}".format(t))
senial_cuadrada_scipy = sig.square(2 * np.pi * frec_senial * t)
print("Señal cuadrada scipy: {}".format(senial_cuadrada_scipy))

plt.plot(t, senial_cuadrada_scipy)
plt.ylim(-2, 2)

######## MAXIMOS Y MINIMOS SCIPY CUADRADA #################

# Encuentra el valor mínimo y sus índices
minimo_valor = np.min(senial_cuadrada_scipy)
indices_minimos = np.where(senial_cuadrada_scipy == minimo_valor)[0]  # De la tupla extrae el indice

# Encontrar el valor máximo y sus índices
maximo_valor = np.max(senial_cuadrada_scipy)
indices_maximos = np.where(senial_cuadrada_scipy == maximo_valor)[0]  # De la tupla extrae el indice

# Mostrar resultados
print("Valor mínimo:", minimo_valor)
print("Índices del valor mínimo:", indices_minimos)

print("Valor máximo:", maximo_valor)
print("Índices del valor máximo:", indices_maximos)

######################################################
```

```python
# 3:  Señal triangular con scipy
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

#Parámetros de la señal
frec_muestreo = 1000 #en Hz
frec_senial = 2 # en Hz
t = np.arange(0,2,1/frec_muestreo)

print("t: {}".format(t))
senial_triangular_scipy = sig.sawtooth(2 * np.pi * frec_senial * t)
print("Señal triangular scipy: {}".format(senial_triangular_scipy))

plt.plot(t, senial_triangular_scipy)
plt.ylim(-1.2, 1.2)

#fig, ax = plt.subplots()
#ax.plot(senial_triangular_scipy,"r", linewidth=2)
#ax.grid(True)
#plt.ylim(-2, 2)
#plt.show()

######## MAXIMOS Y MINIMOS NUMPY TRIANGULAR #################

# Encuentra el valor mínimo y sus índices
minimo_valor = np.min(senial_triangular_scipy)
indices_minimos = np.where(senial_triangular_scipy == minimo_valor)[0]  # De la tupla extrae el indice

# Encontrar el valor máximo y sus índices
maximo_valor = np.max(senial_triangular_scipy)
indices_maximos = np.where(senial_triangular_scipy == maximo_valor)[0]  # De la tupla extrae el indice

# Mostrar resultados
print("Valor mínimo:", minimo_valor)
print("Índices del valor mínimo:", indices_minimos)

print("Valor máximo:", maximo_valor)
print("Índices del valor máximo:", indices_maximos)

######################################################
```

```python
# 3:  Señal senoidal con numpy
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

#Parámetros de la señal
frec_muestreo = 5000 #en Hz
frec_senial = 2 # en Hz
t = np.arange(0,1,1/frec_muestreo)

print("t: {}".format(t))
senial_senoidal_np = np.sin(2 * np.pi * frec_senial * t)
print("Señal senoidal numpy: {}".format(senial_senoidal_np))

plt.plot(t, senial_senoidal_np)
plt.ylim(-1.2, 1.2)

######## MAXIMOS Y MINIMOS NUMPY TRIANGULAR #################

# Encuentra el valor mínimo y sus índices
minimo_valor = np.min(senial_senoidal_np)
indices_minimos = np.where(senial_senoidal_np == minimo_valor)[0]  # De la tupla extrae el indice

# Encontrar el valor máximo y sus índices
maximo_valor = np.max(senial_senoidal_np)
indices_maximos = np.where(senial_senoidal_np == maximo_valor)[0]  # De la tupla extrae el indice

# Mostrar resultados
print("Valor mínimo:", minimo_valor)
print("Índices del valor mínimo:", indices_minimos)

print("Valor máximo:", maximo_valor)
print("Índices del valor máximo:", indices_maximos)

######################################################
```

```python
# 3:  Delta de Dirac con scipy
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

#Parámetros de la señal
frec_muestreo = 32 #en Hz
frec_senial = 2 # en Hz
t = np.arange(0,1,1/frec_muestreo)

print("t: {}".format(t))
Delta_Dirac_scipy = sig.unit_impulse(frec_muestreo, int(frec_muestreo/2))
print("Delta de Dirac scipy: {}".format(Delta_Dirac_scipy))

plt.plot(t, Delta_Dirac_scipy)
plt.ylim(-0.1, 1.1)

######## MAXIMOS Y MINIMOS NUMPY TRIANGULAR #################

# Encuentra el valor mínimo y sus índices
minimo_valor = np.min(Delta_Dirac_scipy)
indices_minimos = np.where(Delta_Dirac_scipy == minimo_valor)[0]  # De la tupla extrae el indice

# Encontrar el valor máximo y sus índices
maximo_valor = np.max(Delta_Dirac_scipy)
indices_maximos = np.where(Delta_Dirac_scipy == maximo_valor)[0]  # De la tupla extrae el indice

# Mostrar resultados
print("Valor mínimo:", minimo_valor)
print("Índices del valor mínimo:", indices_minimos)

print("Valor máximo:", maximo_valor)
print("Índices del valor máximo:", indices_maximos)

######################################################
```

```python
# 4. Calcular máximos y mínimos con subindices en python puro.

min = 9999
max = 0
indice_min = 0
indice_max = 0

array = [5,8,14,1,24,9,48,88,7,33,6,21,19]
print('En Python puro...')
print('Array ingresado:',array)

for i in range(len(array)):
    if array[i] < min:
        min = array[i]
        indice_min = i
    if array[i] > max:
        max = array[i]
        indice_max = i

print('El valor mínimo es:',min,'y su índice es:',indice_min)
print('El valor máximo es:',max,'y su índice es:',indice_max)

# 5. Idem anterior con numpy.

print('\nCon Numpy...')
array2 = np.array(array)
print('Array ingresado:',array2)
print('El valor mínimo es:',array2.min(),'y su índice es:',array2.argmin())
print('El valor máximo es:',array2.max(),'y su índice es:',array2.argmax())

```
